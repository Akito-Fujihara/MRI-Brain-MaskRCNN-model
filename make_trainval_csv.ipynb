```
import os.path as osp
from dataloader import make_datapath_list, BrainDataset, get_transform
import cv2
import pandas as pd
import numpy as np
import csv
```

```
rootpath = osp.abspath("")
makepath = make_datapath_list(rootpath)
train_img_list, train_anno_list = makepath.get_list("train")
val_img_list, val_anno_list = makepath.get_list("val")

print("train dataset:", len(train_img_list),"validation dataset",len(val_img_list))
```

```
import matplotlib.pyplot as plt
%matplotlib inline

def show_image(img):
  plt.imshow(img)
  plt.show()
```

```
train_img_anno_path = []
for img_path, anno_path in zip(train_img_list, train_anno_list):
  anno_img = cv2.imread(anno_path)
  if np.count_nonzero(anno_img == 255) > 0:
    train_img_anno_path.append([img_path, anno_path])
```

```
pd_columns = ["train_img_path", "train_anno_path"]
pd_train_img_anno_path = pd.DataFrame(train_img_anno_path)
pd_train_img_anno_path.columns = pd_columns
pd_train_img_anno_path.to_csv("train_path.csv")
```

```
val_img_anno_path = []
for img_path, anno_path in zip(val_img_list, val_anno_list):
  anno_img = cv2.imread(anno_path)
  if np.count_nonzero(anno_img == 255) > 0:
    val_img_anno_path.append([img_path, anno_path])
```

```
pd_columns = ["val_img_path", "val_anno_path"]
pd_val_img_anno_path = pd.DataFrame(val_img_anno_path)
pd_val_img_anno_path.columns = pd_columns
pd_val_img_anno_path.to_csv("validation_path.csv")
```

```
def make_path_csv(csv_path):
  csv_file = open(csv_path, "r", encoding="ms932", errors="", newline="" )
  rows = csv.reader(csv_file ,delimiter=",", doublequote=True, lineterminator="\r\n", quotechar='"', skipinitialspace=True)
  img_paths = []
  anno_paths = []
  for row in rows:
    img_paths.append(row[1])
    anno_paths.append(row[2])
  return img_paths[1:], anno_paths[1:]
```

```
make_path_csv("validation_path.csv")
```
